```python
import os
import json
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt

from analysis import *
from plots import *

%matplotlib inline

with open('config.json') as f:
    config = json.load(f)
```

```python
result_dir = '../results'

output_dir = config['output_dir']
# output_dir = '../psd_output/'
output_dir = '../v1_2_output/'
```

## Select sessions for analysis

```python
# Read session selection
session_selection_file = 'session_selection'
# session_selection_file = 'session_selection_all'

session_selection = pd.read_csv(os.path.join(result_dir, session_selection_file + '.csv'), index_col='session_id')
```

#### Filter session

```python
genotype = []
session_dir_exist = []
for session_id, row in session_selection.iterrows():
    session_dir_exist.append(os.path.isdir(os.path.join(output_dir, f'session_{session_id:d}')))
    genotype.append(row['full_genotype'].split('/', 1)[0].split('-', 1)[0])

session_selection['session_dir_exist'] = session_dir_exist
session_selection['genotype'] = genotype
session_selection['laser'] = session_selection.index >= 789848216

# seletct optotagged sessions
session_idx = session_selection['session_dir_exist']
session_idx &= (session_selection['genotype'] == 'Pvalb') | (session_selection['genotype'] == 'Sst')
session_selection = session_selection[session_idx].copy()
```

#### Get probe directory

```python
session_ids = []
probe_dirs = []
for session_id, row in session_selection.iterrows():
    session_dir = os.path.join(output_dir, f'session_{session_id:d}')
    if not os.path.isdir(session_dir):
        print(f"Missing session directory for session {session_id:d}")
        continue
    ecephys_structure_acronym = row['ecephys_structure_acronyms']
    info_file = os.path.join(output_dir, f'session_{session_id:d}_{ecephys_structure_acronym:s}.json')
    if os.path.isfile(info_file):
        with open(info_file) as f:
            probe_id = json.load(f)['probe_id']
            probe_dir = os.path.join(session_dir, f'probe_{probe_id:d}')
    else:
        for probe_dir in os.listdir(session_dir):
            if probe_dir.startswith('probe_'):
                probe_dir = os.path.join(session_dir, probe_dir)
                break
        else:
            probe_dir = ''
    if not os.path.isdir(probe_dir):
        print(f"Missing probe directory for session {session_id:d}")
        raise FileNotFoundError(probe_dir)
    session_ids.append(session_id)
    probe_dirs.append(probe_dir)

session_selection = session_selection.loc[session_ids]
session_selection['probe_dir'] = probe_dirs
```

## Detect tagged units

```python
genotypes = ['Pvalb', 'Sst']

# ttest_alpha = {'Pvalb': None, 'Sst': None}  # alpha for t-test
# evoked_ratio_threshold = {'Pvalb': 1.5, 'Sst': 2.0}  # threshold for ratio of evoked to baseline
ttest_alpha = {'Pvalb': 0.05, 'Sst': 0.05}  # alpha for t-test
evoked_ratio_threshold = {'Pvalb': None, 'Sst': None}  # threshold for ratio of evoked to baseline
spike_width_threshold = {'Pvalb': 0.45, 'Sst': None}  # ms. Spike width lower than this are PV
```

```python
optotag_df = []
n_unit_sessions = {}

for session_id, row in session_selection.iterrows():
    probe_dir, ecephys_structure_acronym = row['probe_dir'], row['ecephys_structure_acronyms']
    optotag_file = os.path.join(probe_dir, f'{ecephys_structure_acronym:s}_optotag_units.csv')
    if not os.path.isfile(optotag_file):
        print(f"Missing optotag file for session {session_id:d}")
        continue
    df = pd.read_csv(optotag_file, index_col='unit_id')
    df['session_id'] = session_id
    optotag_df.append(df)
    n_unit_sessions[session_id] = df.shape[0]

optotag_df = pd.concat(optotag_df)  # concatenate units across sessions
```

#### Additional filter

```python
select_session_type = None
# select_session_type = 'functional_connectivity'
# select_session_type = 'brain_observatory_1.1'

select_psd = True
select_laser = False

session_idx = pd.Series(np.ones(session_selection.shape[0], dtype=bool), index=session_selection.index)

if select_session_type is not None:
    session_idx &= session_selection['session_type'] == select_session_type

if select_psd:
    session_idx &= session_selection['psd good']

if select_laser:
    session_idx &= session_selection['laser']
```

```python
optotag_filt_df = {}

for genotype in genotypes:
    idx = (session_selection['genotype'] == genotype) & session_idx
    df = optotag_df.loc[optotag_df['session_id'].isin(idx.index[idx])].copy()

    df, positive_units = detect_optotag(df, ttest_alpha=ttest_alpha[genotype],
        evoked_ratio_threshold=evoked_ratio_threshold[genotype],
        spike_width_threshold=spike_width_threshold[genotype])
    optotag_filt_df[genotype] = df
```

```python
for genotype in genotypes:
    df = optotag_filt_df[genotype]
    print(f"Number of positive units: {sum(df['positive']):d}")

    ax = plot_optotag_units(df, ttest_alpha=ttest_alpha[genotype],
        evoked_ratio_threshold=evoked_ratio_threshold[genotype],
        spike_width_threshold=spike_width_threshold[genotype])
    evoked_ratio = df['evoked_ratio']
    ax.set_ylim(bottom=np.min(evoked_ratio[evoked_ratio > 0]))
    ax.set_yscale('log')
    ax.set_title(genotype)

    plt.show()
```

## Analyze PPC

```python
colors = {'beta': 'b', 'gamma': 'r'}
waves = list(colors)
n_waves = len(waves)
genotype_clrs = {'Pvalb': 'tab:red', 'Sst': 'tab:blue'}

# Set the combination of stimuli
stimulus_names = {
    'brain_observatory_1.1': ['drifting_gratings', 'natural_movie_one', 'natural_movie_three'],
    'functional_connectivity': ['drifting_gratings_75_repeats', 'natural_movie_one_more_repeats', 'natural_movie_one_shuffled']
}

movie_stims = ['natural_movie_one', 'natural_movie_three', 'natural_movie_one_more_repeats', 'natural_movie_one_shuffled']

# Select drifting gratings conditiona
condition_filter = {
    'brain_observatory_1.1': dict(
        cond_name = 'temporal_frequency',
        cond_filt = lambda x: x < 5
    ),
    'functional_connectivity': dict(
        cond_name = 'contrast',
        cond_filt = lambda x: x > 0.5
    )
}
```

```python
n_positive = {}
for genotype, df in optotag_filt_df.items():
    n_positive[genotype] = {}
    for session_type in stimulus_names:
        session_types = session_selection.loc[df['session_id'], 'session_type'].values
        n_positive[genotype][session_type] = sum(df['positive'] & (session_types == session_type))
n_positive = pd.DataFrame.from_dict(n_positive, orient='index')
n_positive.index.name = 'genotype'
print('Number of positive units')
display(n_positive)
```

### Load data and calculate

```python
sorted_channel_id = 2  # second closest channel group to each unit
```

```python
from collections import defaultdict

spike_phases = defaultdict(dict)
for session_id, row in session_selection.iterrows():
    probe_dir = row['probe_dir']
    ecephys_structure_acronym = row['ecephys_structure_acronyms']
    session_type = row['session_type']
    for stim in stimulus_names[session_type]:
        spike_phase_file = os.path.join(probe_dir, f'{ecephys_structure_acronym:s}_{stim:s}_spkie_phase.nc')
        if not os.path.isfile(spike_phase_file):
            print(f"Missing spike phase file for session {session_id:d}")
            continue
        spike_phase = xr.load_dataset(spike_phase_file)
        # recover complex values
        spike_phase = spike_phase.assign(resultant_phase=spike_phase.resultant_phase_real + \
            1j * spike_phase.resultant_phase_imag).drop_vars(['resultant_phase_real', 'resultant_phase_imag'])
        # filter drifting grating conditions
        if stim not in movie_stims:
            cond_filt = condition_filter[session_type]
            presentation_ids = cond_filt['cond_filt'](spike_phase[cond_filt['cond_name']])
            spike_phase = spike_phase.sel(presentation=presentation_ids)
            spike_phase = spike_phase.sel(presentation_id=spike_phase.presentation_ids.values.ravel())
            # spike_phase = spike_phase.assign_attrs(time_window=(0., 2.0))  # set time window for debugging
        # Add to list
        if 'sorted_channel' in spike_phase.resultant_phase.dims:
            spike_phase = spike_phase.sel(sorted_channel=sorted_channel_id)
        spike_phases[stim][session_id] = spike_phase
```

```python
ppc_all = {}  # in whole duration for each stimulus
ppc_windows = {}  # in windows for each stimulus
for genotype, df in optotag_filt_df.items():
    ppc = defaultdict(list)
    ppc_win = defaultdict(lambda: defaultdict(list))
    for stim, spike_phase in spike_phases.items():
        for session_id, spk_pha in spike_phase.items():  # over sessions
            if session_selection.loc[session_id, 'genotype'] != genotype:
                continue
            unit_ids = [i for i in spk_pha.unit_id.values if i in df.index and df.loc[i, 'positive']]  # only positive units
            if not len(unit_ids):
                continue
            # whole duration
            time_window = np.array(spk_pha.attrs['time_window']) / spk_pha.window.size
            ppc[stim].append(phase_locking_value(spk_pha, unit_ids=unit_ids,
                presentation_ids=None, time_window=time_window, unbiased=True))
            # window durations
            if spk_pha.window.size > 1:
                for win in spk_pha.window.values:
                    spk_pha_win = spk_pha.sel(window=win)
                    ppc_win[stim][win].append(phase_locking_value(spk_pha,
                        unit_ids=unit_ids, presentation_ids=spk_pha_win.presentation_ids.values,
                        time_window=spk_pha_win.windows.values, unbiased=True))
    # concatenate across sessions
    ppc_all[genotype] = {stim: xr.concat(x, dim='unit_id') for stim, x in ppc.items()}
    ppc_windows[genotype] = {stim: {win: xr.concat(y, dim='unit_id') for win, y in x.items()} \
                             for stim, x in ppc_win.items()}
```

### Plot PPC

#### PPC vs. mean FR (beta/gamma)

```python
figsize = (3.6, 3)

for genotype in genotypes:
    print(genotype)
    for stim, ppc in ppc_all[genotype].items():
        ppc_waves = []
        for w in waves:
            p = ppc.sel(wave=w)
            ppc_waves.append([p.mean_firing_rate, p.PPC])
        ppc_waves = np.array(ppc_waves)  # (wave, xy, units)

        _, ax = plt.subplots(1, 1, figsize=figsize)
        for cu in ppc_waves.swapaxes(0, -1):  # (units, xy, wave)
            ax.plot(*cu, color='gray', alpha=0.5)  # connect same unit
        for i, w in enumerate(waves):
            ax.plot(*ppc_waves[i], 'o', color=colors[w],
                    markerfacecolor='none', label=w)
        ax.set_xlabel("Mean Firing Rate (Hz)")
        ax.set_ylabel("PPC")
        ax.set_title(f'{stim:s} all')
        ax.legend(loc='upper right', framealpha=0.2)
        plt.show()
```

#### PPC vs. mean FR (duration windows)

```python
win_markers = ['x', 'o', '^']

for genotype in genotypes:
    print(genotype)
    for stim, ppc_win in ppc_windows[genotype].items():
        ppc_wins = []
        for win, ppc in ppc_win.items():
            ppc_waves = []
            for w in waves:
                p = ppc.sel(wave=w)
                ppc_waves.append([p.mean_firing_rate, p.PPC])
            ppc_wins.append(ppc_waves)
        ppc_wins = np.array(ppc_wins).swapaxes(0, 1)  # (wave, windows, xy, units)

        fig, axs = plt.subplots(1, n_waves, figsize=(n_waves * figsize[0], figsize[1]))
        for i, w in enumerate(waves):
            ax = axs[i]
            for cu in ppc_wins[i].swapaxes(0, -1):  # (units, xy, windows)
                ax.plot(*cu, color='gray', alpha=0.5)  # connect same unit
            for j, win in enumerate(ppc_win):
                ax.plot(*ppc_wins[i, j], linestyle='none', marker=win_markers[j],
                        color=colors[w], markerfacecolor='none', label=f'{win:s}')
            ax.set_xlabel("Mean Firing Rate (Hz)")
            ax.set_ylabel("PPC")
            ax.set_title(w)
            ax.legend(loc='upper right', framealpha=0.2)
        fig.suptitle(stim)
        plt.show()
```

```python
ppc_all_df = []
ppc_wins_df = []

for genotype in genotypes:
    for w in waves:
        for stim, ppc in ppc_all[genotype].items():
            p = ppc.sel(wave=w)
            df = pd.DataFrame(dict(unit_id=p.unit_id.values, PPC=p.PPC.values,
                spike_number=p.spike_number.values, mean_firing_rate=p.mean_firing_rate.values,
                wave=w, stimulus=stim, genotype=genotype))
            ppc_all_df.append(df)
        for stim, ppc_win in ppc_windows[genotype].items():
            for win, ppc in ppc_win.items():
                p = ppc.sel(wave=w)
                df = pd.DataFrame(dict(unit_id=p.unit_id.values, PPC=p.PPC.values,
                    spike_number=p.spike_number.values, mean_firing_rate=p.mean_firing_rate.values,
                    wave=w, window=f'{stim:s}\n{win:s}', stimulus=stim, genotype=genotype))
                ppc_wins_df.append(df)
ppc_all_df = pd.concat(ppc_all_df, ignore_index=True)
ppc_wins_df = pd.concat(ppc_wins_df, ignore_index=True)
```

#### Firing rate distribution

```python
df = ppc_wins_df[ppc_wins_df['wave'] == waves[0]]
stim_wins = df['window'].unique()

figsize = (6.4, 2.4)
fig, axs = plt.subplots(len(stim_wins), 1, figsize=(figsize[0], len(stim_wins) * figsize[1]), sharex=True, sharey=True, squeeze=False)
for ax, stim_win in zip(axs.ravel(), stim_wins):
    win_df = df[df['window'] == stim_win]
    frs = win_df['mean_firing_rate'].values
    bins = np.logspace(np.log10(np.min(frs)), np.log10(np.max(frs)), 20)
    idx = [win_df['genotype'] == genotype for genotype in genotypes]
    ax.hist([frs[i] for i in idx], bins=bins, label=genotypes, stacked=True)
    ax.set_xscale('log')
    ax.set_ylabel(stim_win)
    ax.xaxis.set_tick_params(labelbottom=True)
axs[-1, 0].set_xlabel("Mean Firing Rate (Hz)")
axs[0, 0].legend(loc='upper right', framealpha=0.2)
fig.tight_layout()
plt.show()
```

### List average PPC

```python
# select units with high enough firing rate in all stimulus
sel_units = {session_type: set(ppc_all_df['unit_id']) for session_type in stimulus_names}
for session_type, stims in stimulus_names.items():
    for stim in stims:
        sel_units[session_type].intersection_update(ppc_all_df.loc[ppc_all_df['stimulus'] == stim, 'unit_id'])
sel_units = set.union(*sel_units.values())
```

```python
n_positive = {}
for genotype, df in optotag_filt_df.items():
    n_positive[genotype] = {}
    df = df.loc[df.index.isin(sel_units)]
    for session_type in stimulus_names:
        session_types = session_selection.loc[df['session_id'], 'session_type'].values
        n_positive[genotype][session_type] = sum(df['positive'] & (session_types == session_type))
n_positive = pd.DataFrame.from_dict(n_positive, orient='index')
n_positive.index.name = 'genotype'
print('Number of selected units')
display(n_positive)
```

```python
stim_windows = []
ppc_df = defaultdict(dict)
for genotype in genotypes:
    for stim, ppc in ppc_all[genotype].items():
        # get different durations of each stimulus
        ppcs = []
        stim_wins = []
        stim_win = f'{stim:s} whole'
        if stim_win not in stim_windows:
            stim_windows.append(stim_win)
        ppcs.append(ppc)
        stim_wins.append(stim_win)
        if stim in ppc_windows[genotype]:
            for win, ppc in ppc_windows[genotype][stim].items():
                stim_win = f'{stim:s} {win:s}'
                if stim_win not in stim_windows:
                    stim_windows.append(stim_win)
                ppcs.append(ppc)
                stim_wins.append(stim_win)
        for stim_win, ppc in zip(stim_wins, ppcs):
            ppc = ppc.sel(unit_id=[i in sel_units for i in ppc.unit_id.values])
            w = waves[0]  # firing rate same for all waves
            ppc_df[(genotype, 'mean firing rate')][stim_win] = ppc.mean_firing_rate.sel(wave=w).mean().item()
            ppc_df[(genotype, 'std firing rate')][stim_win] = ppc.mean_firing_rate.sel(wave=w).std().item()
            for w in waves:
                ppc_df[(genotype, 'PPC ' + w)][stim_win] = ppc.sel(wave=w).PPC.mean().item()

ppc_df = pd.DataFrame(ppc_df, index=pd.Index(stim_windows, name='stimulus duration'),
                      columns=pd.MultiIndex.from_tuples(list(ppc_df)))

with pd.option_context('display.float_format', '{:.3g}'.format, 'display.max_columns', None):
    for genotype in genotypes:
        print(ppc_df.loc[:, (genotype, slice(None))])

# ppc_df.to_csv(os.path.join(result_dir, 'entrainment.csv'))
```
